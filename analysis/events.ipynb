```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rc("font", family="Malgun Gothic")
plt.rc('axes', unicode_minus=False)
plt.rcParams['grid.color'] = (0.5, 0.5, 0.5, 0.1)
df = pd.read_csv("../.out/events.csv")
df
```

```python
plt.figure(figsize=(20, 8))

x_label = "N(GP/PA)"
y_label = "수익률"
plt.subplot(2, 2, 1)
plt.grid(True)
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.scatter(x=df[x_label], y=df[y_label], alpha=0.2)

x_label = "R(GP/PA)"
plt.subplot(2, 2, 2)
plt.grid(True)
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.scatter(x=df[x_label], y=df[y_label], alpha=0.2)

plt.subplot(2, 2, 3)
plt.grid(True)
revenue_by_rank = pd.read_csv("../.out/revenue_by_factor_rank.csv")
plt.bar(revenue_by_rank[revenue_by_rank.columns[0]], revenue_by_rank["AVG(수익률)"])

plt.show()
```
